## STA examples

```python
# some packages used in this example
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

### Example of 1-dim stimuls

```python
# load stimulus (1-dim)
stim=np.loadtxt('ex1/stim.txt')
# load spike vector 
spike=np.loadtxt('ex1/spike.txt')
```

```python
# Let's plot them
plt.subplot(211)
plt.plot(stim)
plt.ylabel('stim')
plt.subplot(212)
plt.plot(spike, 'r')
plt.ylabel('spike')
plt.xlabel('t')
# Sometimes, we have more than one spikes in a bin. But, that's not a big problem.
```

```python
## find stimulus pattern that triggers a spike

# Step 1. find spike timing (in terms of index)
spikeTime = []
for t, sp in enumerate(spike):
    if sp >  0 and t > W:   # keep time greater than the window size
        #print(t)
        spikeTime.append(t)
#print(spikeTime)

# Step 2. grap stim patterns that triggered spikes
W = 50;    # window size
stimTriggeredSpike = []
for t in spikeTime:
    stimTriggeredSpike.append(stim[t-W:t])
stimTriggeredSpike = np.matrix(stimTriggeredSpike)    
print(stimTriggeredSpike.shape)

# calc mean across patterns to get STA
STA = stimTriggeredSpike.mean(0).transpose()  
plt.plot(STA)
plt.xlabel('t')
plt.ylabel('STA')
```

### Example of 2-dim stimulus

```python
# load stimulus (1-dim)
stim=np.loadtxt('ex3/stim.txt')
# load spike vector 
spike=np.loadtxt('ex3/spike.txt')
```

```python
T, dim = stim.shape # 40000 x 8
print(T,dim)
```

```python
## find stimulus pattern that triggers a spike

# Step 1. find spike timing (in terms of index)
spikeTime = []
for t, sp in enumerate(spike):
    if sp >  0 and t > W:   # keep time greater than the window size
        #print(t)
        spikeTime.append(t)
#print(len(spikeTime))


# Step 2. grap stim patterns that triggered spikes

W = 50;    # window size

stimTriggeredSpike = []
for t in spikeTime:
    stimTriggeredSpike.append(stim[t-W:t,])
print(np.asarray(stimTriggeredSpike).shape)

# calc mean across patterns to get STA
STA = np.asarray(stimTriggeredSpike).mean(0)
plt.imshow(STA.transpose())
plt.xlabel('t')
plt.ylabel('ch')
plt.title('STA')
```
